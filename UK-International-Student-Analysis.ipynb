```python
import pandas as pd

# Real number from Hesa 2023/23 reports

data = {
      "Country": ["China", "India", "Nigeria", "USA", "Pakistan",
                   "Hong Kong", "Malaysia", "Kenya", "Bangladesh", "Ghana"],
      "Students_2023": [151430, 126520, 49985, 16790, 22875,
                        24635, 18920, 11240, 14350, 12870],
      "Students_2022": [143960, 108815, 39885, 15995, 18020,
                        25560, 18290, 9545, 12480, 10545]
}

df = pd.DataFrame(data)
print(df)
```

```python
df = pd.DataFrame(data)

plt.figure(figsize=(12,6 ))
sns.barplot(data=df, x="Country", y="Students_2023", palette="Blues_d")
plt.title("International Students in the UK by Country (2023)", fontsize=16)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.barplot(data=df, x="Country", y="Students_2023", hue="Country", palette="Blues_d", legend=False)

plt.title("International Students in the Uk by Country(2023)", fontsize=16)

plt.xlabel("Country of Orogin")

plt.ylabel("Number of Students")

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python

df["Growth"] = df["Students_2023"] - df["Students_2022"]

plt.figure(figsize=(12,6))
sns.barplot(data=df, x="Country", y="Growth", hue="Country", palette="Greens_d", legend=False)
plt.title("Growth in International Students by Country (2022 vs 2023)", fontsize=16)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Students")
plt.xticks(roration=45)
plt.tight_layout()
plt.show()
```

```python
deadline_data = {
    "University": ["Manchester", "UCL", "Edinburgh", "Birmingham", "Leeds",
                   "Nottingham", "Sheffield", "Bristol", "Cardiff", "Liverpool",
                    "Sunderland"],
    
    "Total_International_Students": [12500, 11800, 9600, 8400, 7900,
                                     7200, 6800, 6500, 5400, 5100,
                                     3200],

    "Met_Deadline": [7875, 7670, 6240, 5460, 5135,
                     4680, 4420, 4225, 3510, 3315,
                     1920],

    "Requested_Extension": [2500, 2360, 1920, 1680, 1580,
                            1440, 1360, 1300, 1080, 1020,
                            640],

    "Defaulted": [2125, 1770, 1440, 1260, 1185,
                  1080, 1020, 975, 810, 765,
                  640]
}

df_deadline = pd.DataFrame(deadline_data)
print(df_deadline)
```

```python
# Payment deadline chart:

df_deadline_melted = df_deadline.melt (
    id_vars="University",
    value_vars= ["Met_Deadline", "Requested_Extension", "Defaulted"],
    var_name= "Payment_status",
    value_name= "Students"
)

plt.figure(figsize=(14, 7))
sns.barplot(data=df_deadline_melted, 
x="University", y ="Students", hue="Payment_status", palette="RdYlGn")
plt.title("International Students Payment Outcomes by University", fontsize=14)
plt.xlabel("University")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
deadline_data = {
      "University": ["Manchester", "UCL", "Edinburgh", "Birmingham", "Leeds",
                     "Nottingham", "Sheffield", "Bristol", "Cardiff", "Liverpool",
                     "Sunderland"],
      "Total_International": [12500, 11800, 9600, 8400, 7900,
                              7200, 6800, 6500, 5400, 5100,
                              3200],
      "Met_Deadline": [7875, 7670, 6240, 5460, 5135,
                       4680, 4420, 4225, 3510, 3315,
                       1920],
      "Requested_Extension": [2500, 2360, 1920, 1680, 1580,
                              1440, 1360, 1300, 1080, 1020,
                              640],
      "Defaulted": [2125, 1770, 1440, 1260, 1185,
                    1080, 1020, 975, 810, 765,
                    640]
  }

df_deadline = pd.DataFrame(deadline_data)

df_deadline["Default_Rate"] = (df_deadline["Defaulted"] / 
                               df_deadline["Total_International"]) * 100

df_deadline["Extension_Rate"] = (df_deadline["Requested_Extension"] /
                                 df_deadline["Total_International"]) * 100

df_deadline["OnTime_Rate"] = (df_deadline["Met_Deadline"] / 
                              df_deadline["Total_International"]) * 100

print(df_deadline[["University", "OnTime_Rate", "Extension_Rate", "Default_Rate"]])
```

```python
df_rates_melted = df_deadline.melt(                                                                     id_vars="University",
      value_vars=["OnTime_Rate", "Extension_Rate", "Default_Rate"],                                       var_name="Payment_Status",
      value_name="Percentage"
)

plt.figure(figsize=(14, 7))
sns.barplot(data=df_rates_melted, x="University", y="Percentage", hue="Payment_Status",
palette="RdYlGn")
plt.title("International Student Payment Rates by University (%)", fontsize=14)
plt.xlabel("University")
plt.ylabel("Percentage of Students (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
graduate_data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    "Graduate_Visas_Issued": [7100, 56900, 107800, 125300, 138600],
    "Sponsored_Jobs_Found": [2840, 19915, 35574, 43855, 48510]

}

df_graduate = pd.DataFrame(graduate_data)
print(df_graduate)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_graduate ["Year"], df_graduate["Graduate_Visas_Issued"],
    marker="o", color="blue", label="Graduate Visas Issued")

plt.plot(df_graduate ["Year"], df_graduate["Sponsored_Jobs_Found"],
    marker="o", color="green", label="Sponsored Jobs Found")

plt.title("Graduate Route Visas vs Spnonsored Jobs Found (2021-2025)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Students")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#what percentage of visa holders actually got sponsored jobs each year

df_graduate["Success_Rate"] = (df_graduate["Sponsored_Jobs_Found"] / 
                               df_graduate["Graduate_Visas_Issued"]) * 100

plt.figure(figsize=(10, 6))
    
plt.plot(df_graduate["Year"], df_graduate["Success_Rate"],
         marker="o", color="red", label="Sponsorship Success Rate (%)")

plt.title("Percentage of Graduate Visa Holders Who Found Sponsored Jobs (2021-2025)",fontsize=12)

plt.xlabel("Year")
plt.ylabel("Success Rate (%)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
recommendations = {
    "Area": ["Payment Flexibility", "Payment Flexibility", "Payment Flexibility",
               "Sponsorship Support", "Sponsorship Support", "Sponsorship Support",
               "Student Wellbeing", "Student Wellbeing"],
    
    "Recommendation": ["Extend installment period until graduation",
                         "Never block Canvas or portal for missed payments",
                         "Create emergency hardship fund",
                         "Partner with sponsors before students graduate",
                         "Run on-campus sponsored job fairs each semester",
                         "Provide CV coaching for sponsorship roles",
                         "Send payment reminders 2 weeks before deadline",
                         "Assign financial advisor to every international student"],
    
     "Impact": ["High", "High", "Medium", "High", "High", "Medium", "Medium", "High"],
    
     "Effort": ["Medium", "Low", "Medium", "High", "Medium", "Low", "Low", "Medium"]
  }

df_recommendations = pd.DataFrame(recommendations)
print(df_recommendations)
```

```python
plt.figure(figsize=(12, 8))

colors_impact = df_recommendations["Impact"].map({"High": "green", "Medium": "orange", "Low":
  "red"})

plt.barh(df_recommendations["Short_Label"], df_recommendations["Impact_Score"],
           color=colors_impact)

plt.title("Recommendations Ranked by Impact on International Students", fontsize=13)
plt.xlabel("Impact Score (1=Low, 2=Medium, 3=High)")
plt.ylabel("Recommendation")
plt.xlim(0, 4)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 10))

colors = {"Payment Flexibility": "blue", "Sponsorship Support": "green", "Student Wellbeing":
  "orange"}

positions = {
      "Never block Canvas":    (1.0, 3.0),
      "Extend installments":   (1.5, 2.5),
      "Hardship fund":         (2.0, 2.0),
      "Partner with sponsors": (3.0, 3.0),
      "Job fairs on campus":   (2.5, 3.0),
      "CV coaching":           (1.0, 2.0),
      "Payment reminders":     (1.2, 1.5),
      "Financial advisor":     (2.0, 3.0)
  }

df_recommendations["Pos_X"] = df_recommendations["Short_Label"].map(lambda x: positions[x][0])
df_recommendations["Pos_Y"] = df_recommendations["Short_Label"].map(lambda x: positions[x][1])

for area, group in df_recommendations.groupby("Area"):
      ax.scatter(group["Pos_X"], group["Pos_Y"],
                 label=area, color=colors[area], s=800, zorder=5)

for i, row in df_recommendations.iterrows():
      ax.text(row["Pos_X"], row["Pos_Y"] - 0.15,
              row["Short_Label"],
              fontsize=12, ha="center", va="top", fontweight="bold")

ax.set_title("Recommendations — Impact vs Effort", fontsize=12)
ax.set_xlabel("Effort Required", fontsize=12)
ax.set_ylabel("Impact on Students", fontsize=12)
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(["Low", "Medium", "High"])
ax.set_yticks([1, 2, 3])
ax.set_yticklabels(["Low", "Medium", "High"])
ax.set_xlim(0.5, 3.8)
ax.set_ylim(0.8, 3.6)
ax.legend(loc="lower right")
plt.tight_layout()
plt.show()
```
